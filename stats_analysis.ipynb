### Variáveis de controle de refinamento dos modificadores

```python
version = 'v2'
```

```python
import numpy as np
import matplotlib.pyplot as plt

means = np.array([df_gk['Overall'].mean(), df_cen['Overall'].mean(), df_win['Overall'].mean(),
                  df_cen['Overall'].mean(), df_str['Overall'].mean()])
positions = ('Goleiros', 'Zagueiros', 'Laterais', 'Meio-campistas', 'Atacantes')
pos_qnt = np.arange(len(positions))

fig = plt.figure()
plt.bar(pos_qnt, means)
plt.xticks(pos_qnt, positions)
plt.title('Overall Means')
plt.show()
fig.savefig('imgs/overall/overall_means_' + version + '.png')
```
